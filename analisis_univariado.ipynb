## Preparación del dataset

```python
print("\nValores nulos:")
df.isna().sum()
```

```python
print("\nFilas duplicadas:")
df.duplicated().sum()
```

```python
# Identificar outliers
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[ ((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR))) ]
    return outliers

columnas_numericas = df.select_dtypes(include=np.number).columns.tolist()

for columna in columnas_numericas:
    outliers_en_columna = find_outliers_IQR(df[columna])
    print(f"--- Outliers en la columna: {columna} ---")
    print(f'Número de outliers: {len(outliers_en_columna)}')
    if len(outliers_en_columna) > 0:
        print(f'Max valor outlier: {outliers_en_columna.max()}')
        print(f'Min valor outlier: {outliers_en_columna.min()}')

    print("\nOutliers del conjunto de datos:")
    print("\n")
```

```python
#Convertir una variable cualitativa en cuantitativa para calcular la correlación entre las variables
colum_cuantitativa = ['Position']
# Aplica la codificación One-Hot
df_codificado = pd.get_dummies(df, columns=colum_cuantitativa, dtype=int)

df_codificado.info()
```

```python
print("\nMapa de calor:")
matriz_corr = df_codificado.corr()
plt.figure(figsize=(12,10))
sns.heatmap(matriz_corr, annot=True, fmt=".1f", cmap="coolwarm", annot_kws={"size": 10})
plt.show()
```
